#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

```python
#code

def std_dev(dataframe, column):
    
    df=dataframe
    series = df[column]
    mew = np.mean(series)
    n = len(series)-1
    xminusmewsq= [(xi-mew)**2 for xi in series]
    summation_sq=np.sum(xminusmewsq)
    standard_dev=(summation_sq/n)**0.5
    
    return standard_dev
```

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

```python
#Do any states show have >50% participation on *both* tests either year?

#we create a function that uses boolean masking to help us determine this:

def above_50(column1, column2):
    boolean = (final[column1]>50) & (final[column2]>50)
    return final[boolean][['state', column1, column2]]
```

```python
above_50('2017_act_participation', '2017_sat_participation')
```

```python
above_50('2018_act_participation', '2018_sat_participation')
```

Answer: We see that in 2017, Florida, Georgia and Hawaii had > 50% participation in both tests. This list then expanded to include North and South Carolina in 2018.
